```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def create_exact_journal_figure():
    # 1. Setup the Canvas (High Resolution, Pure White Background)
    fig, ax = plt.subplots(figsize=(20, 6), dpi=300)
    fig.patch.set_facecolor('white')
    ax.set_facecolor('white')
    ax.set_xlim(0, 36)
    ax.set_ylim(0, 10.5)
    ax.axis('off') # Hide the grid and axes

    # --- Colors (Frontier AI Journal Palette) ---
    c_bg = '#F8FAFC'      # Very light slate/blue for professional blocks
    c_edge = '#1E293B'    # Dark slate for crisp borders
    c_biz_bg = '#166534'  # Deep green for final output
    c_biz_txt = 'white'

    # --- Helper Functions for Exact Drawing ---
    def draw_box(x, y, w, h, text, facecolor=c_bg, text_color='black', font_weight='normal', text_size=10):
        # Using FancyBboxPatch for very subtly rounded, modern corners
        rect = patches.FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0,rounding_size=0.15",
                                      linewidth=1.2, edgecolor=c_edge, facecolor=facecolor, zorder=2)
        ax.add_patch(rect)
        ax.text(x + w/2, y + h/2, text, ha='center', va='center', fontsize=text_size,
                color=text_color, fontweight=font_weight, zorder=3, linespacing=1.4)

    def draw_complex_box(x, y, w, h, title, bullets, facecolor=c_bg, text_color='black'):
        rect = patches.FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0,rounding_size=0.15",
                                      linewidth=1.2, edgecolor=c_edge, facecolor=facecolor, zorder=2)
        ax.add_patch(rect)
        # Title at the top center
        ax.text(x + w/2, y + h - 0.4, title, ha='center', va='top', fontsize=10, fontweight='bold', color=text_color, zorder=3)
        # Bullets aligned to the left
        ax.text(x + 0.3, y + h - 1.6, bullets, ha='left', va='top', fontsize=9.5, color=text_color, zorder=3, linespacing=1.6)

    def draw_arrow(x1, y1, x2, y2):
        ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                    arrowprops=dict(arrowstyle="-|>", mutation_scale=12, color=c_edge, lw=1.2), zorder=1)

    def draw_bracket(x_start, x_end, y_base, text):
        y_top = y_base + 0.2
        y_mid = y_top + 0.2
        x_mid = (x_start + x_end) / 2
        # Lines forming the bracket { rotated horizontally
        ax.plot([x_start, x_start], [y_base, y_top], color=c_edge, lw=1.2)
        ax.plot([x_end, x_end], [y_base, y_top], color=c_edge, lw=1.2)
        ax.plot([x_start, x_mid], [y_top, y_top], color=c_edge, lw=1.2)
        ax.plot([x_mid, x_end], [y_top, y_top], color=c_edge, lw=1.2)
        ax.plot([x_mid, x_mid], [y_top, y_mid], color=c_edge, lw=1.2)
        # Bracket Text
        ax.text(x_mid, y_mid + 0.15, text, ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

    # --- 2. Draw Nodes (Exact Coordinates) ---
    y_center = 4.5  # Central alignment axis

    # Data Layer
    draw_box(0.5, 2.5, 3.0, 4.0, "Real Data\nCollection\n\n(n = 2,138)\n\nSurvey (Online\n+ Offline,\nBangladesh)", font_weight='bold')
    draw_complex_box(4.5, 2.5, 3.5, 4.0, "Synthetic Data\nGeneration", "• Distribution\n  Preservation\n• Correlation\n  Consistency\n• Gaussian Noise\n  Injection")
    draw_box(9.0, 3.5, 2.5, 2.0, "Hybrid Dataset\n(=20,751\nsamples)", font_weight='bold')

    # Feature Engineering
    draw_complex_box(12.5, 2.0, 4.0, 5.0, "Behavioral Feature\nConstruction", "• UTAUT Constructs\n  (PE, EE, PR, FC)\n• Aggregation\n  (Mean Scores)\n• Interaction\n  Features (PE * FC,\n  EE * PR, PR * FC)")

    # Predictive Modeling
    draw_box(17.5, 6.0, 3.5, 1.2, "Logistic\nRegression (Most\nCost-Efficient\nModel)")
    draw_box(17.5, 4.5, 3.5, 0.8, "Deep Learning")
    draw_box(17.5, 3.2, 3.5, 0.8, "XGBoost")
    draw_box(17.5, 1.9, 3.5, 0.8, "Random Forest")
    draw_box(22.5, 3.25, 2.5, 2.5, "Churn\nPrediction\n(Binary\nOutput)", font_weight='bold')

    # Economic Evaluation
    draw_complex_box(26.0, 2.0, 4.0, 5.0, "Economic Impact\nAnalysis", r"• Customer Lifetime" + "\n" + r"  Value (CLV)" + "\n" + r"• Retention Gain ($\Delta$)" + "\n" + r"• Cost-Efficiency (" + "\n" + r"  $\eta = G / C_{model}$)")

    # Business Output
    draw_box(31.5, 3.5, 2.5, 2.0, "Business\nDecision\nSupport", facecolor=c_biz_bg, text_color=c_biz_txt, font_weight='bold', text_size=11)

    # --- 3. Draw Connecting Arrows ---
    # Horizontal flow lines
    draw_arrow(3.5, y_center, 4.5, y_center)
    draw_arrow(8.0, y_center, 9.0, y_center)
    draw_arrow(11.5, y_center, 12.5, y_center)

    # Feature Engineering -> Models (Splitter)
    ax.plot([16.5, 17.0], [y_center, y_center], color=c_edge, lw=1.2, zorder=1) # Out from Feat
    ax.plot([17.0, 17.0], [2.3, 6.6], color=c_edge, lw=1.2, zorder=1)           # Vertical distributor
    draw_arrow(17.0, 6.6, 17.5, 6.6) # To Logistic
    draw_arrow(17.0, 4.9, 17.5, 4.9) # To DL
    draw_arrow(17.0, 3.6, 17.5, 3.6) # To XGBoost
    draw_arrow(17.0, 2.3, 17.5, 2.3) # To Random Forest

    # Models -> Churn (Collector)
    ax.plot([21.0, 21.75], [6.6, 6.6], color=c_edge, lw=1.2, zorder=1)
    ax.plot([21.0, 21.75], [4.9, 4.9], color=c_edge, lw=1.2, zorder=1)
    ax.plot([21.0, 21.75], [3.6, 3.6], color=c_edge, lw=1.2, zorder=1)
    ax.plot([21.0, 21.75], [2.3, 2.3], color=c_edge, lw=1.2, zorder=1)
    ax.plot([21.75, 21.75], [2.3, 6.6], color=c_edge, lw=1.2, zorder=1)         # Vertical collector
    draw_arrow(21.75, y_center, 22.5, y_center)                                 # Into Churn

    # Final straight connections
    draw_arrow(25.0, y_center, 26.0, y_center)
    draw_arrow(30.0, y_center, 31.5, y_center)

    # --- 4. Draw Top Brackets ---
    y_bracket = 7.5
    draw_bracket(0.5, 11.5, y_bracket, "Data layer")
    draw_bracket(12.5, 16.5, y_bracket, "Feature Engineering")
    draw_bracket(17.5, 25.0, y_bracket, "Predictive Modeling")
    draw_bracket(26.0, 30.0, y_bracket, "Economic Evaluation")

    # --- 5. Save & Render ---
    plt.tight_layout()
    plt.savefig("framework_exact_figure.png", dpi=400, bbox_inches='tight', transparent=False)
    plt.savefig("framework_exact_figure.pdf", bbox_inches='tight')
    print("Exact matched diagram generated successfully as 'framework_exact_figure.png' and '.pdf'")

if __name__ == "__main__":
    create_exact_journal_figure()
```
